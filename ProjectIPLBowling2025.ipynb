```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_bowlers= pd.read_csv('IPL2025Bowlers.csv')

#pd.set_option('display.max_rows',1000)
df_bowlers
```

# Purple Cap LeaderBoard

##### Top 10 

```python
# Take top 10 bowlers
top_bowlers = df_bowlers.head(10)
top_bowlers.sort_values(['WKT'],ascending=False,inplace=True)

plt.figure()

plt.bar(top_bowlers['Player Name'], top_bowlers['WKT'])

plt.xticks(rotation=45)
plt.xlabel("Player Name")
plt.ylabel("Wickets")
plt.title("Top 10 Bowlers by Wickets (IPL 2025)")

plt.tight_layout()
plt.show()
```

# Selecting Quality Bowlers of IPL 2025

###### Bowlers who have taken more than 10 wickets and played atleast 7 matches in an IPL Season can be considered as a QUality Bowler

```python
df_bowlers=df_bowlers.loc[(df_bowlers['WKT']>=10)&(df_bowlers['INN']>=7)]
df_bowlers.sort_values(['WKT'],ascending=False,inplace=True)
df_bowlers
```

```python
team_counts= df_bowlers['Team'].value_counts()
plt.figure()

team_counts.plot(kind='bar')

plt.xlabel("Team")
plt.ylabel("Number of Bowlers")
plt.title("Number of Quality Bowlers per Team (IPL 2025)")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# Selecting Wicket-Takers of IPL 2025

###### Bowlers who on an average take 1.2 Wickets per match and satisfying the Quality Bowlers condition is a Wicket-Taker Bowler

```python
df_bowlers=pd.read_csv('IPL2025Bowlers.csv')
df_bowlers['WPM']=df_bowlers['WKT']/df_bowlers['INN']
df_bowlers
```

```python
df_bowlers=df_bowlers.loc[(((df_bowlers['WKT']>=10)&(df_bowlers['INN']>=7))&(df_bowlers['WPM']>1.2))]
df_bowlers.sort_values(['WPM'],ascending=False,inplace=True)
df_bowlers
```

```python
df_bowlers['rank']=range(1,len(df_bowlers)+1)
df_bowlers.set_index('rank')
```

```python
top_10=df_bowlers.head(10)
plt.figure()
plt.bar(top_10['Player Name'],top_10['WPM'])

plt.xticks(rotation=45)
plt.xlabel("Player Name")
plt.ylabel("WPM")
plt.title("Top 10 Wicket Takers (IPL 2025)")

plt.tight_layout()
plt.show()

```

```python
team_counts= df_bowlers['Team'].value_counts()
plt.figure()
team_counts.plot(kind='bar')
plt.xlabel("Team")
plt.ylabel("Number of Bowlers")
plt.title("Distribution of Bowlers by Team (IPL 2025)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

# Economical Bowlers

##### Bowled at least 25 overs and economy less than 8.5

```python
df_bowlers=pd.read_csv('IPL2025Bowlers.csv')
df_bowlers['Bowls']=(df_bowlers['RUNS']/df_bowlers['ECO'])*6
df_bowlers['Bowls']=df_bowlers['Bowls'].round().astype(int)
df_bowlers
```

```python
df_bowlers=df_bowlers.loc[((df_bowlers['ECO']<=8.5)&(df_bowlers['Bowls']>=150))]
df_bowlers.sort_values(['ECO'],ascending=True,inplace=True)
df_bowlers['rank']=range(1,len(df_bowlers)+1)
df_bowlers.set_index('rank')
df_bowlers
```

```python
plt.figure()
plt.bar(df_bowlers['Player Name'],df_bowlers['ECO'])
plt.xticks(rotation=45)
plt.xlabel("Player Name")
plt.ylabel("Economy")
plt.title("Economy Rates of Bowlers (IPL 2025)")
plt.tight_layout()
plt.show()
```

```python
team_counts= df_bowlers['Team'].value_counts()
plt.figure()
plt.bar(df_bowlers['Team'],team_counts)
plt.xticks(rotation=45)
plt.xlabel("Team Name")
plt.ylabel("Number of Players")
plt.title("Distribution of Bowlers by Team (IPL 2025)")
plt.tight_layout()
plt.show()
```
