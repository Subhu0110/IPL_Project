```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Purple Cap LeaderBoard

##### Top 10

# Selecting Quality Bowlers of IPL 2025

###### Bowlers who have taken more than 10 wickets and played atleast 7 matches in an IPL Season can be considered as a QUality Bowler

```python
df_bowlers=df_bowlers[(df_bowlers['WKT']>=10)&(df_bowlers['INN']>=7)]
df_bowlers.sort_values(['WKT'],ascending=False,inplace=True)
df_bowlers
```

# Selecting Wicket-Takers of IPL 2025

###### Bowlers who on an average take 1.2 Wickets per match and satisfying the Quality Bowlers condition is a Wicket-Taker Bowler

```python
df_bowlers=pd.read_csv('IPL2025Bowlers.csv')
df_bowlers['WPM']=df_bowlers['WKT']/df_bowlers['INN']
df_bowlers
```

```python
df_bowlers=df_bowlers.loc[(((df_bowlers['WKT']>=10)&(df_bowlers['INN']>=7))&(df_bowlers['WPM']>1.2))]
df_bowlers.sort_values(['WPM'],ascending=False,inplace=True)
df_bowlers
```

```python
df_bowlers['rank']=range(1,len(df_bowlers)+1)
df_bowlers.set_index('rank',inplace=True)
df_bowlers
```

```python
top_10=df_bowlers.head(10)
plt.figure()
plt.bar(top_10['Player Name'],top_10['WPM'])

plt.xticks(rotation=45)
plt.xlabel("Player Name")
plt.ylabel("WPM")
plt.title("Top 10 Wicket Takers (IPL 2025)")

plt.tight_layout()
plt.show()

```

```python
team_counts= df_bowlers['Team'].value_counts()
team_counts['RR']=0
plt.figure()
team_counts.plot(kind='bar')
plt.xlabel("Team")
plt.ylabel("Number of Bowlers")
plt.title("Distribution of Bowlers by Team (IPL 2025)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

# Economical Bowlers

##### Bowled at least 25 overs and economy less than 8.5

```python
df_bowlers=pd.read_csv('IPL2025Bowlers.csv')
df_bowlers['Bowls']=(df_bowlers['RUNS']/df_bowlers['ECO'])*6
df_bowlers['Bowls']=df_bowlers['Bowls'].round().astype(int)
df_bowlers
```

```python
df_bowlers=df_bowlers.loc[((df_bowlers['ECO']<=8.5)&(df_bowlers['Bowls']>=150))]
df_bowlers.sort_values(['ECO'],ascending=True,inplace=True)
df_bowlers['rank']=range(1,len(df_bowlers)+1)
df_bowlers.set_index('rank',inplace=True)
df_bowlers
```

```python
plt.figure()
plt.bar(df_bowlers['Player Name'],df_bowlers['ECO'])
plt.xticks(rotation=45)
plt.xlabel("Player Name")
plt.ylabel("Economy")
plt.title("Economy Rates of Bowlers (IPL 2025)")
plt.tight_layout()
plt.show()
```

```python
team_counts= df_bowlers['Team'].value_counts()
plt.figure()
team_counts.plot(kind='bar')
plt.xticks(rotation=45)
plt.xlabel("Team Name")
plt.ylabel("Number of Players")
plt.title("Distribution of Economical Bowlers by Team (IPL 2025)")
plt.tight_layout()
plt.show()
```

# Most Valuable Bowler

```python
df_bowlers=pd.read_csv('IPL2025Bowlers.csv')
```

```python
df_bowlers['WKT_norm'] = (df_bowlers['WKT'] - df_bowlers['WKT'].min()) / (df_bowlers['WKT'].max() - df_bowlers['WKT'].min())
df_bowlers['ECO_norm'] = (df_bowlers['ECO'].max() - df_bowlers['ECO']) / (df_bowlers['ECO'].max() - df_bowlers['ECO'].min())
df_bowlers['AVG_norm'] = (df_bowlers['AVG'].max() - df_bowlers['AVG']) / (df_bowlers['AVG'].max() - df_bowlers['AVG'].min())
df_bowlers['SR_norm']  = (df_bowlers['SR'].max() - df_bowlers['SR']) / (df_bowlers['SR'].max() - df_bowlers['SR'].min())
```

```python
df_bowlers['SCORE']= df_bowlers['WKT_norm']*0.25+df_bowlers['ECO_norm']*0.25+df_bowlers['AVG_norm']*0.25+df_bowlers['SR_norm']*0.25
df_bowlers.sort_values('SCORE',ascending=False,inplace=True)
df_bowlers['rank']=range(1,len(df_bowlers)+1)
df_bowlers.set_index('rank',inplace=True)
```

```python
df_bowlers.head(10)
```

```python
df_mvp=df_bowlers.head(10)
plt.figure()
plt.bar(df_mvp['Player Name'],df_mvp['SCORE'])
plt.xlabel('Player Name')
plt.ylabel('MVP Score')
plt.xticks(rotation=45)
plt.title('Most Valuable Bowlers IPL (2025)')
plt.tight_layout()
plt.show()
```
